```python
#Tokenization
```

```python
#Sentence Tokenize

from nltk import sent_tokenize, word_tokenize

import nltk
nltk.download('punkt')
```

```python
sent_tok =[]

for sent in clean_text_1:
    sent = sent_tokenize(sent)
    sent_tok.append(sent)
```

```python
#words tokenization
```

```python
#we will use List Comprehension
```

```python
clean_text_2 =[]
```

```python
clean_text_2 =[word_tokenize(i) for i in clean_text_1]
```

```python
clean_text_2 
```

```python
#Removing Punctuation & Special Character
```

```python
import re
```

```python
clean_text_3 =[]

for words in clean_text_2:
    clean =[]
    for w in words:
        res = re.sub(r'[^/w/s]', "", w)
        if res !=  "":
            clean.append(res)
        clean_text_3.append(clean)
```

```python
clean_text_3
```

```python
#STOPWORDS Removal
```

```python
#downloading stopwords

import nltk 
nltk.download('stopwords')
```

```python
from nltk.corpus import stopwords
```

```python
clean_text_4 =[]

for words in clean_text_3:
    w=[]
    
    for word in words:
        if not word in stopwords.words('english'):
            w.append(word)
    
    clean_text_4.append(w)
        
    
    
```
